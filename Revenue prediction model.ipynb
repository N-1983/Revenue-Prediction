```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
model1 = DecisionTreeRegressor()
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
from sklearn.ensemble import RandomForestRegressor
model3 = RandomForestRegressor()

model_pipeline = [model1, model2, model3]
model_names = ['Regression Tree', 'Linear Regression', 'Random Forest Regression']
scores = {}
i=0
for model in model_pipeline:
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=10))
    scores[model_names[i]] = mean_score
    i = i+1
scores_table = pd.DataFrame([scores])
scores_table = scores_table.T
scores_table.columns = ['Cross Validation Score']
scores_table = scores_table.sort_values(by = 'Cross Validation Score', ascending=False)
```

```python
print('\n\nPredicting with Linear Regression and Random Forest Regressor')
print(100*'_')
print("\n")
model2 = model2.fit(X_train, y_train)
prediction2020_LR = []
for i in model2.predict(X_train):
    prediction2020_LR.append(i)
predict_table_LR = pd.DataFrame(prediction2020_LR)
predict_table_LR.columns = ['Revenue_2020 (Predicted - LR)']
model3 = model3.fit(X_train, y_train)
prediction2020_RF = []
for i in model3.predict(X_train):
    prediction2020_RF.append(i)
predict_table_RF = pd.DataFrame(prediction2020_RF)
predict_table_RF.columns = ['Revenue_2020 (Predicted - RF)']
predict_table = pd.concat([predict_table_LR,predict_table_RF],axis=1)
display(predict_table)
```
